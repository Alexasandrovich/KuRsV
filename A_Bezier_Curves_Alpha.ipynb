```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
partition = np.linspace(0, 1, 100)
```

```python
dots_x = []
dots_y = []
def Curve(p, q, r, t):
    coordinate = (1 - t)**2 * p + 2*t*(1 - t)*q + t**2 * r
    return coordinate
```

```python
for t in partition:
    dots_x.append(Curve(40, 30, 11, t))
    dots_y.append(Curve(30, 31, 15, t))
```

```python
scatter1 = plt.plot(dots_x, dots_y)
```
